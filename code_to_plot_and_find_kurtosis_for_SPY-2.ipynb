```python
# import modules
from datetime import datetime
import yfinance as yf
import matplotlib.pyplot as plt
import numpy as np
# initialize parameters
start_date = datetime(2000, 1, 1)
end_date = datetime(2021, 1, 1)

# get the data
data = yf.download('GOOGL', start = start_date,
				end = end_date)

data


returns = np.log(data['Adj Close']) - np.log(data['Adj Close'].shift(1))
print(returns)
import seaborn as sn #to get the distribution curve
import matplotlib.pyplot as plt #to plot the graph
import pandas as pd #to get the data in a desirable format



sn.displot(returns.dropna(),bins=70,color='black', height=3,aspect=23/5, kde=True)
plt.show()


res = pd.Series(returns,name="Range")
plot = sn.distplot(res,rug=True,hist=False)
plt.show()


from scipy.stats import kurtosis
import scipy
kurtosis(returns.dropna())
```
