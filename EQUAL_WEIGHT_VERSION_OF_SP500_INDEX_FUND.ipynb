Alternative / More Efficient

```python
def chunks(lst, n):
    """Yield successive n-sized chunks from lst."""
    for i in range(0, len(lst), n):
        yield lst[i:i + n]
```

```python
symbgroups = list(chunks(stocks['Symbol'],100 ))
symbgroups
symbstrings=[]
for i in range(0,len(symbgroups)):
    symbstrings.append(','.join(symbgroups[i]))

tablefinall = pd.DataFrame(columns= ['Ticker','Stock Price','Market Cap','Number of Shares to Buy'])


for symbstring in symbstrings:
    batchapicallurl = f'https://sandbox.iexapis.com/stable/stock/market/batch?symbols={symbstring}&types=quote&token={IEX_CLOUD_API_TOKEN}'
    print(batchapicallurl)
    data= requests.get(batchapicallurl).json()
    for symbol in symbstring.split(','):
        tablefinall =tablefinall.append(
            pd.Series(
            [
                 symbol,
                 data[symbol]['quote']['latestPrice'],
                 data[symbol]['quote']['marketCap'],
                 'na'
            ],
            index= columns),
            ignore_index=True
         )

tablefinall
```

```python
bgcolor = '#0a029'
fontclr = '#ffffff'

stringformat = w.book.add_format(
    { 'font_color':fontclr,
      'bg_color':bgcolor,
      'border': 1} )
dollarformat = w.book.add_format(
    { 'num_format': '$0.00',
      'font_color':fontclr,
      'bg_color':bgcolor,
      'border': 1})
integerformat = w.book.add_format(
    { 'num_format': '0',
      'font_color':fontclr,
      'bg_color':bgcolor,
      'border': 1})
```
